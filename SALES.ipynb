```python
import pandas as pd

df = pd.read_csv(r"C:\Users\karth\Downloads\Sales.csv")
print(df.shape)
print(df.columns)
```

```python
print(df.isnull().sum())

df[df.select_dtypes(include=['number']).columns] = df.select_dtypes(include=['number']).fillna(0)

df[df.select_dtypes(include=['object']).columns] = df.select_dtypes(include=['object']).fillna("Unknown")

df[df.select_dtypes(include=['datetime']).columns] = df.select_dtypes(include=['datetime']).fillna(pd.Timestamp("2000-01-01"))

print("\nAfter Filling")
print(df.tail())
```

```python
df['Revenue_per_Item'] = df['Sales'] / df['Quantity']
print(df[['Product', 'Sales', 'Quantity', 'Revenue_per_Item']])
```

```python
df['Category'] = df['Sales'].apply(lambda x: "High" if x>1000 else "Low")
df['RegionCode'] = df['Region'].map({'North':'N','South':'S','East':'E','West':'W'})
df['Quantity'] = df['Quantity'].apply(lambda x: x+5)
print(df[['Region', 'RegionCode', 'Sales', 'Category', 'Quantity']])

```

```python
print(df[df['Product'].str.lower().str.contains('laptop')])
```

```python
df = df.drop_duplicates()
print(df.shape)
print(df)
```

```python
pivot = pd.pivot_table(df, values='Sales',
                    index='Region', 
                  columns='Product', 
                   aggfunc='sum')
print(pivot)
```
